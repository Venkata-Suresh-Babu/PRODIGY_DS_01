```python
#Bar chart
worldpopulation_df['2023'].plot(kind='hist', bins=20, title='2023')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.gca().spines[['top', 'right',]].set_visible(False)
plt.show()
```

```python
#Line chart
worldpopulation_df['2023'].plot(kind='line', figsize=(8, 4), title='2023')
plt.xlabel("Year")
plt.ylabel("Population")
plt.gca().spines[['top', 'right']].set_visible(False)
plt.show()
```
